```python
from prophet import Prophet
```

```python
df = df[['date', 'sales', 'onpromotion']]
```

```python
# Rename columns for Prophet
df = df.rename(columns={
    'date': 'ds',
    'sales': 'y'
})
```

```python
# Sort by date
df = df.sort_values('ds')

# Keep last 90 days for testing
train = df.iloc[:-90]
test = df.iloc[-90:]
```

```python
len(test)
```

```python
len(train)
```

```python
model = Prophet(
    yearly_seasonality=True,
    weekly_seasonality=True,
    daily_seasonality=False
)

# Add extra regressor
model.add_regressor('onpromotion')

# Fit model
model.fit(train)
```

```python
# Forecast next 30 days
future = model.make_future_dataframe(periods=30)

# Add onpromotion (assume no promotion in future)
future['onpromotion'] = 0
```

```python
future
```

```python
forecast = model.predict(future)

# Show predictions
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail())
```

```python
forecast_output = forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
forecast_output.to_csv("future_sales_forecast.csv", index=False)

print("Future forecast saved successfully!")
```
